```python
import pandas as pd
import numpy as np
```

```python
dataset = 'rte'
df = pd.read_csv(f'{dataset}_results.tex', sep='&', header=None, skiprows=7, engine='python', skipfooter=4)
# remove leading and trailing whitespace and \\
df = df.apply(lambda x: x.str.strip().str.replace(r'\\', ''))
# get the column names by reading the first row of the table, remove \textbf{}
df.columns = df.iloc[0].str.replace(r'\textbf{', '').str.replace('}', '')

# remove the first row
df = df.iloc[1:]

df['Shots'] = df['Shots'].str.replace(r'\multirow{5}{*}{', '').str.strip('}')

# remove the Sample column
df = df.drop('Sample', axis=1)

# remove \textbf{ } and strip for {} except for Method column
for col in df.columns:
    if col != 'Method':
        df[col] = df[col].str.replace(r'\textbf{', '').str.replace('}', '').str.strip('{ \t')
# df = df.apply(lambda x: x.str.replace(r'\textbf{', '').str.replace('}', '').str.strip('{ \t'))

df = df.replace('', np.nan)
# fill na for shots as ffill
df['Shots'] = df['Shots'].fillna(method='ffill')
# fill empty cells with NaN

df
```

```python
# fillna for Shots with ffill
df['Shots'] = df['Shots'].fillna(method='ffill')
# groupyby shots and method and output as latex table
# print(sst2.set_index(['Shots', 'Method']).to_latex())
hightlighted_sst2 = df.copy()
hightlighted_sst2 = hightlighted_sst2.set_index(['Shots', 'Method'])
hightlighted_sst2
```

```python
def highlight_max(s):
    # print(s)
    # print(s.name)
    if s.name == 'Method':
        return s
    else:
        # split by '±' and take the first element
        modified_s = s.str.split('±').str[0].astype(float) 
        print(modified_s)
        is_max = modified_s == modified_s.max()
        is_second_max = modified_s == modified_s.nlargest(2).iloc[-1]
        # print(is_min)
        s[is_max] = s[is_max].apply(lambda x: f'\\textbf{{{x}}}')
        # s[is_second_max] = s[is_second_max].apply(lambda x: f'\\underline{{{x}}}')
    return s

results = []
# fill nan by 0
hightlighted_sst2 = hightlighted_sst2.fillna('0')

for shot in hightlighted_sst2.index.get_level_values('Shots').unique():
    r = hightlighted_sst2.loc[shot, :].apply(highlight_max, axis=0)
    r['Shots'] = shot
    r = r.reset_index()
    r = r.set_index(['Shots', 'Method'])
    results.append(r)
```

```python
results_df = pd.concat(results)
# replace all nan by '-'
results_df = results_df.fillna('-')
# round the results by 2 dp
results_df = results_df.round(2)
# output results_df to latex with all latex command keep

# for all entry that contains ±, do split and [0]_$\pm [1]$
results_df = results_df.applymap(lambda x: f'{x.split("±")[0].strip()}$_{{\pm{x.split("±")[1].strip("} ")}}}$' + ('}' if x[-1] == '}' else '') if (len(str(x).split("±")) == 2) else x)
# replace \underline{\textbf{0}} by ''
# results_df = results_df.replace('\\underline{\\textbf{0}}', '')
results_df = results_df.replace('\\textbf{0}', '')

# for shots, add $ in front and back
results_df = results_df.reset_index()
# results_df['Shots'] = '$' + results_df['Shots'].astype(str) + '$'
results_df = results_df.set_index(['Shots', 'Method'])

# results_df = results_df.style.format(precision=2)
# to latex
results_df.to_latex(f'{dataset}_final_results.tex', escape=False)
results_df
```
